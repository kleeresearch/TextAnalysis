```python
from gensim.models import Word2Vec
```

```python
with open('2016_filtered_review.txt', encoding='utf-8') as f:
    docs = [line.strip().split('\t\t') for line in f]
```

```python
docs[:5]
```

```python
docs1 = [doc[1].strip().split() for doc in docs]
```

```python
print(docs1[:5])
```

윈도우 크기는 3, 타겟 단어를 중심으로 양옆에 3개 단어를 문맥단어로 사용하겠다는 것을 의미.

단어의 최소 빈도는 3개. 최소 3번 이상 사용된 단어들만 사용하겠다.

벡터사이즈는 생성하는 각각의 벡터의 차원의 크기를 이야기한다.

벡터가 100개의 우너소를 가진다.

각각의 차원의 크기가 작다고 간주한다. 저차원 벡터를 embedding vector라고 표현한다.

특정한 단어에 벡터를 확인하기 위해서는 객체가 가지고 있는 wb라는 변수를 벡터의 정보를 얻고자 하는 문자열 key 정보를 넣으면 된다.

리턴이 되는 원소의 수를 출력한다. 주변의 단어가 유사한 단어들이 유사한 벡터를 가진다. 원소의 값이 유사하다는 것이 주요한 내용이다.


특정 단어에서 벡터 정보를 추출한 다음에 단어들 간의 유사도를 파악할 수 있다. 벡터간의 유사도를 파악할 수 있다. 위치 정보를 파악할 수 있다. 방향 정보를 기반으로 유사도를 파악할 수 있다.

```python
model = Word2Vec(docs1, window=3, min_count=3, vector_size=100)
#window => 보고자 하는 양쪽 단
#min_count => 전체 corpus에서 사용 빈도수가 min_count보다 낮은 단어 제외
#size => word vector의 크기
# https://radimrehurek.com/gensim/models/word2vec.html
```

객체를 만들기 위해서 word2vec 함수를 만든다.

```python
len(model.wv['이정재'])
```

```python
model.wv['이정재']
```

```python
import numpy as np
```

```python
print(np.dot(model.wv['이정재'], model.wv['이범수'])/(np.linalg.norm(model.wv['이정재'])*np.linalg.norm(model.wv['이범수'])))
```

이정재와 이범수라는 코사인 셋타의 값이다. 그 값이 0.79로 나왔다.

gensim에서 제공이 되는 similar_by_world 펑션은 특정 단어와 유사도를 기준으로 유사도가 큰 상위 50개 단어를 출력할 수 있다.


벡터들 간의 코사인 유사도 - 이 단어들이 비슷한 문맥에서 사용이 되었기 때문.

```python
model.wv.similar_by_word('이정재',topn=50)
```

```python
model.wv.similar_by_word('올드보이',topn=20)
```

```python
model.wv.similar_by_word('최악',topn=50)
```

```python
model.wv.similarity(u'이정재', u'이범수')
```
